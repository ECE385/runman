```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
f = 440
bitwidth = 16
time = 1/440

# 1 second @ 44.1 KHz steps
time = np.arange(0, time, 1/44_100)

# sine (f=f) samples
samples = np.sin(2 * np.pi * f * time)

# discretize to bitwidth
discrete = np.rint(samples * (2**(bitwidth-1) - 1)).astype(np.int16)

plt.plot(time[:4000], samples[:4000])
plt.plot(time[:4000], discrete[:4000])
plt.plot(time[:4000], discrete.astype(np.uint16)[:4000])



```

```python
# Save file
FILE_NAME = "sin.mem"
np.savetxt(FILE_NAME, discrete.astype(np.uint16), fmt="%04x")
```
